```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import cv2
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import random

# Set memory growth to avoid OOM errors
physical_devices = tf.config.list_physical_devices('GPU')
if physical_devices:
    for device in physical_devices:
        tf.config.experimental.set_memory_growth(device, True)
    print(f"GPU is available: {len(physical_devices)} devices found")
else:
    print("No GPU found, using CPU mode")
    # Set TensorFlow to use CPU only
    os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

# Set random seeds for reproducibility
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

class BrainTumorClassifier:
    def __init__(self, img_height=128, img_width=128, batch_size=8):
        """
        Initialize the Brain Tumor Classifier
        
        Parameters:
        -----------
        img_height : int
            Height to resize input images (reduced to 128x128)
        img_width : int
            Width to resize input images (reduced to 128x128)
        batch_size : int
            Batch size for training (reduced to 8)
        """
        self.img_height = img_height
        self.img_width = img_width
        self.batch_size = batch_size
        self.n_classes = 4  # glioma, meningioma, no_tumor, pituitary
        self.class_names = ['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']
        
        # Initialize model
        self.model = None
        
    def load_data(self, training_dir, testing_dir, max_samples_per_class=None):
        """
        Load segmented images from training and testing directories
        
        Parameters:
        -----------
        training_dir : str
            Directory containing training images
        testing_dir : str
            Directory containing testing images
        max_samples_per_class : int or None
            Maximum number of samples to load per class (to limit dataset size)
            
        Returns:
        --------
        X_train, y_train, X_test, y_test : Training and testing data
        """
        X_train = []
        y_train = []
        X_test = []
        y_test = []
        
        # Load training data
        print("Loading training data...")
        for class_idx, class_name in enumerate(self.class_names):
            class_dir = os.path.join(training_dir, class_name)
            if not os.path.exists(class_dir):
                print(f"Warning: {class_dir} not found!")
                continue
            
            files = os.listdir(class_dir)
            # Filter segmented images
            segmented_files = [f for f in files if f.endswith("_segmented_result.png")]
            
            # Limit the number of samples if specified
            if max_samples_per_class:
                if len(segmented_files) > max_samples_per_class:
                    segmented_files = random.sample(segmented_files, max_samples_per_class)
                    print(f"Limited {class_name} training samples to {max_samples_per_class}")
            
            for file in tqdm(segmented_files):
                img_path = os.path.join(class_dir, file)
                img = cv2.imread(img_path)
                if img is None:
                    continue
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = cv2.resize(img, (self.img_width, self.img_height))
                img = img / 255.0  # Normalize to [0, 1]
                X_train.append(img)
                y_train.append(class_idx)
        
        # Load testing data
        print("Loading testing data...")
        for class_idx, class_name in enumerate(self.class_names):
            class_dir = os.path.join(testing_dir, class_name)
            if not os.path.exists(class_dir):
                print(f"Warning: {class_dir} not found!")
                continue
            
            files = os.listdir(class_dir)
            # Filter segmented images 
            segmented_files = [f for f in files if f.endswith("_segmented_result.png")]
            
            # Limit the number of samples if specified
            if max_samples_per_class:
                test_max = max(10, max_samples_per_class // 4)  # Ensure at least 10 test samples
                if len(segmented_files) > test_max:
                    segmented_files = random.sample(segmented_files, test_max)
                    print(f"Limited {class_name} testing samples to {test_max}")
            
            for file in tqdm(segmented_files):
                img_path = os.path.join(class_dir, file)
                img = cv2.imread(img_path)
                if img is None:
                    continue
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = cv2.resize(img, (self.img_width, self.img_height))
                img = img / 255.0  # Normalize to [0, 1]
                X_test.append(img)
                y_test.append(class_idx)
                    
        # Convert to numpy arrays
        X_train = np.array(X_train)
        y_train = np.array(y_train)
        X_test = np.array(X_test)
        y_test = np.array(y_test)
        
        # Convert labels to one-hot encoding
        y_train = to_categorical(y_train, num_classes=self.n_classes)
        y_test = to_categorical(y_test, num_classes=self.n_classes)
        
        print(f"Training data shape: {X_train.shape}")
        print(f"Training labels shape: {y_train.shape}")
        print(f"Testing data shape: {X_test.shape}")
        print(f"Testing labels shape: {y_test.shape}")
        
        # Check if we have enough data
        if len(X_train) < 10 or len(X_test) < 10:
            print("Warning: Very small dataset detected! Results may be unreliable.")
            
        return X_train, y_train, X_test, y_test
    
    def build_minimal_cnn_model(self):
        """
        Build a very lightweight CNN model for brain tumor classification
        """
        model = Sequential([
            # First convolutional block - very small number of filters
            Conv2D(16, (3, 3), activation='relu', padding='same', 
                   input_shape=(self.img_height, self.img_width, 3)),
            MaxPooling2D(pool_size=(2, 2)),
            
            # Second convolutional block
            Conv2D(32, (3, 3), activation='relu', padding='same'),
            MaxPooling2D(pool_size=(2, 2)),
            
            # Third convolutional block (fewer filters than original)
            Conv2D(64, (3, 3), activation='relu', padding='same'),
            MaxPooling2D(pool_size=(2, 2)),
            Dropout(0.3),
            
            # Classification layers
            Flatten(),
            Dense(128, activation='relu'),
            Dropout(0.5),
            Dense(self.n_classes, activation='softmax')
        ])
        
        # Compile model
        model.compile(
            optimizer=Adam(learning_rate=0.0005),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        # Print model summary
        model.summary()
        
        self.model = model
        return model
    
    def create_simple_dataset(self, X, y, augment=False):
        """
        Create a batch generator for training (avoids using tf.data which can consume memory)
        
        Parameters:
        -----------
        X : np.array
            Image data
        y : np.array
            Labels
        augment : bool
            Whether to apply simple data augmentation
            
        Returns:
        --------
        generator function
        """
        n_samples = len(X)
        indices = np.arange(n_samples)
        
        # Define a generator function
        def batch_generator():
            while True:
                np.random.shuffle(indices)
                
                for start_idx in range(0, n_samples, self.batch_size):
                    batch_indices = indices[start_idx:start_idx + self.batch_size]
                    batch_X = X[batch_indices].copy()
                    batch_y = y[batch_indices].copy()
                    
                    # Simple augmentation if requested
                    if augment:
                        for i in range(len(batch_X)):
                            # Random horizontal flip with 50% probability
                            if np.random.rand() > 0.5:
                                batch_X[i] = np.fliplr(batch_X[i])
                    
                    yield batch_X, batch_y
                    
        # Calculate steps per epoch
        steps = int(np.ceil(n_samples / self.batch_size))
        
        return batch_generator(), steps
    
    def train(self, X_train, y_train, X_val=None, y_val=None, epochs=15, augment=True):
        """
        Train the CNN model with reduced complexity for CPU training
        
        Parameters:
        -----------
        X_train : np.array
            Training images
        y_train : np.array
            Training labels
        X_val : np.array
            Validation images
        y_val : np.array
            Validation labels
        epochs : int
            Number of training epochs (reduced from 50 to 15)
        augment : bool
            Whether to apply data augmentation
            
        Returns:
        --------
        history : Training history
        """
        if self.model is None:
            self.build_minimal_cnn_model()
            
        # If validation data not provided, split from training data
        if X_val is None or y_val is None:
            X_train, X_val, y_train, y_val = train_test_split(
                X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
            )
        
        # Create generators
        train_gen, train_steps = self.create_simple_dataset(X_train, y_train, augment=augment)
        val_gen, val_steps = self.create_simple_dataset(X_val, y_val, augment=False)
        
        # Define callbacks with more aggressive early stopping
        callbacks = [
            ModelCheckpoint(
                'brain_tumor_classifier_minimal.h5',
                save_best_only=True,
                monitor='val_accuracy',
                mode='max'
            ),
            EarlyStopping(
                patience=3,  # Very short patience to stop early
                monitor='val_loss',
                restore_best_weights=True
            ),
            ReduceLROnPlateau(
                factor=0.2,
                patience=2,  # Reduce learning rate quickly
                min_lr=1e-6,
                monitor='val_loss'
            )
        ]
        
        # Train model
        print("Training minimal CNN model...")
        history = self.model.fit(
            train_gen,
            steps_per_epoch=train_steps,
            validation_data=val_gen,
            validation_steps=val_steps,
            epochs=epochs,
            callbacks=callbacks
            # workers and use_multiprocessing parameters removed
        )
        
        self.history = history
        return history
        
    def evaluate(self, X_test, y_test):
        """
        Evaluate the trained model
        
        Parameters:
        -----------
        X_test : np.array
            Test images
        y_test : np.array
            Test labels
            
        Returns:
        --------
        results : dict
            Evaluation results
        """
        if self.model is None:
            print("Error: Model not trained yet!")
            return None
        
        # For CPU-friendly evaluation, use smaller batches
        batch_size = min(8, len(X_test))
        
        # Evaluate model
        print("Evaluating model...")
        loss, accuracy = self.model.evaluate(X_test, y_test, batch_size=batch_size, verbose=1)
        print(f"Test Loss: {loss}")
        print(f"Test Accuracy: {accuracy}")
        
        # Get predictions (in smaller batches to avoid memory issues)
        y_pred_all = []
        batch_size = min(16, len(X_test))
        for i in range(0, len(X_test), batch_size):
            batch_end = min(i + batch_size, len(X_test))
            batch_pred = self.model.predict(X_test[i:batch_end], verbose=0)
            y_pred_all.append(batch_pred)
        
        y_pred = np.vstack(y_pred_all)
        y_pred_classes = np.argmax(y_pred, axis=1)
        y_true_classes = np.argmax(y_test, axis=1)
        
        # Classification report
        print("\nClassification Report:")
        class_report = classification_report(
            y_true_classes, 
            y_pred_classes, 
            target_names=self.class_names,
            output_dict=True
        )
        print(classification_report(y_true_classes, y_pred_classes, target_names=self.class_names))
        
        # Confusion matrix
        print("\nConfusion Matrix:")
        conf_mat = confusion_matrix(y_true_classes, y_pred_classes)
        print(conf_mat)
        
        try:
            # Plot confusion matrix
            plt.figure(figsize=(10, 8))
            sns.heatmap(
                conf_mat, 
                annot=True, 
                fmt='d', 
                cmap='Blues', 
                xticklabels=self.class_names,
                yticklabels=self.class_names
            )
            plt.xlabel('Predicted')
            plt.ylabel('True')
            plt.title('Confusion Matrix')
            plt.tight_layout()
            plt.savefig('confusion_matrix.png')
            plt.show()
            
            # Plot training history if available
            if hasattr(self, 'history'):
                plt.figure(figsize=(12, 5))
                
                plt.subplot(1, 2, 1)
                plt.plot(self.history.history['accuracy'])
                plt.plot(self.history.history['val_accuracy'])
                plt.title('Model Accuracy')
                plt.ylabel('Accuracy')
                plt.xlabel('Epoch')
                plt.legend(['Train', 'Validation'], loc='upper left')
                
                plt.subplot(1, 2, 2)
                plt.plot(self.history.history['loss'])
                plt.plot(self.history.history['val_loss'])
                plt.title('Model Loss')
                plt.ylabel('Loss')
                plt.xlabel('Epoch')
                plt.legend(['Train', 'Validation'], loc='upper left')
                
                plt.tight_layout()
                plt.savefig('training_history.png')
                plt.show()
                
            # Sample predictions (only a few to save memory)
            self.visualize_predictions(X_test, y_true_classes, y_pred_classes, num_samples=3)
        except Exception as e:
            print(f"Warning: Could not create visualizations: {e}")
        
        return {
            'accuracy': accuracy,
            'loss': loss,
            'classification_report': class_report,
            'confusion_matrix': conf_mat
        }
    
    def visualize_predictions(self, X_test, y_true, y_pred, num_samples=3):
        """
        Visualize some sample predictions
        
        Parameters:
        -----------
        X_test : np.array
            Test images
        y_true : np.array
            True class indices
        y_pred : np.array
            Predicted class indices
        num_samples : int
            Number of samples to visualize (reduced to 3)
        """
        # Select random samples
        indices = np.random.choice(range(len(X_test)), size=min(num_samples, len(X_test)), replace=False)
        
        plt.figure(figsize=(12, 3 * num_samples))
        for i, idx in enumerate(indices):
            plt.subplot(num_samples, 1, i+1)
            plt.imshow(X_test[idx])
            if y_true[idx] == y_pred[idx]:
                color = 'green'
            else:
                color = 'red'
            plt.title(f"True: {self.class_names[y_true[idx]]}, Predicted: {self.class_names[y_pred[idx]]}", color=color)
            plt.axis('off')
        
        plt.tight_layout()
        plt.savefig('sample_predictions.png')
        plt.show()
    
    def predict(self, img_path):
        """
        Make prediction on a single image
        
        Parameters:
        -----------
        img_path : str
            Path to the image
            
        Returns:
        --------
        prediction : dict
            Prediction results including class and confidence
        """
        if self.model is None:
            print("Error: Model not trained yet!")
            return None
            
        # Read and preprocess image
        img = cv2.imread(img_path)
        if img is None:
            print(f"Error: Could not read image at {img_path}")
            return None
            
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (self.img_width, self.img_height))
        img = img / 255.0  # Normalize to [0, 1]
        
        # Make prediction
        pred = self.model.predict(np.expand_dims(img, axis=0))[0]
        pred_class = np.argmax(pred)
        
        try:
            # Show image and prediction
            plt.figure(figsize=(6, 6))
            plt.imshow(img)
            plt.title(f"Predicted: {self.class_names[pred_class]} ({pred[pred_class]:.2f})")
            plt.axis('off')
            plt.show()
        except Exception as e:
            print(f"Warning: Could not display image: {e}")
        
        return {
            'class': self.class_names[pred_class],
            'confidence': float(pred[pred_class]),
            'all_probabilities': {class_name: float(p) for class_name, p in zip(self.class_names, pred)}
        }
    
    def save_model(self, path='brain_tumor_classifier_minimal.h5'):
        """
        Save the trained model
        
        Parameters:
        -----------
        path : str
            Path to save the model
        """
        if self.model is None:
            print("Error: Model not trained yet!")
            return
            
        self.model.save(path)
        print(f"Model saved to {path}")
    
    def load_model(self, path='brain_tumor_classifier_minimal_v1_1.h5'):
        """
        Load a trained model
        
        Parameters:
        -----------
        path : str
            Path to the saved model
        """
        self.model = tf.keras.models.load_model(path)
        print(f"Model loaded from {path}")


def main():
    # Define paths to segmented data
    base_dir = "segmented_mri_dataset"  # Change this to your segmented data directory
    train_dir = os.path.join(base_dir, "Training")
    test_dir = os.path.join(base_dir, "Testing")
    
    # Initialize classifier with smaller image size
    classifier = BrainTumorClassifier(img_height=128, img_width=128, batch_size=16)
    
    # Load data with limit to reduce dataset size for CPU training
    # Adjust max_samples_per_class based on your computer's memory capacity
    max_samples = 300  # Limit samples per class to make training feasible on CPU
    X_train, y_train, X_test, y_test = classifier.load_data(
        train_dir, test_dir, max_samples_per_class=max_samples
    )
    
    # Build and train minimal CNN model
    classifier.build_minimal_cnn_model()
    history = classifier.train(X_train, y_train, epochs=50, augment=True)
    
    # Evaluate model
    results = classifier.evaluate(X_test, y_test)
    
    # Save model
    classifier.save_model()
    
    print("Done!")


if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import cv2
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import random

# Set memory growth to avoid OOM errors
physical_devices = tf.config.list_physical_devices('GPU')
if physical_devices:
    for device in physical_devices:
        tf.config.experimental.set_memory_growth(device, True)
    print(f"GPU is available: {len(physical_devices)} devices found")
else:
    print("No GPU found, using CPU mode")
    # Set TensorFlow to use CPU only
    os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

# Set random seeds for reproducibility
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

class BrainTumorClassifier:
    def __init__(self, img_height=128, img_width=128, batch_size=8):
        """
        Initialize the Brain Tumor Classifier
        
        Parameters:
        -----------
        img_height : int
            Height to resize input images (reduced to 128x128)
        img_width : int
            Width to resize input images (reduced to 128x128)
        batch_size : int
            Batch size for training (reduced to 8)
        """
        self.img_height = img_height
        self.img_width = img_width
        self.batch_size = batch_size
        self.n_classes = 4  # glioma, meningioma, no_tumor, pituitary
        self.class_names = ['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']
        
        # Initialize model
        self.model = None
        
    def load_data(self, training_dir, testing_dir, max_samples_per_class=None):
        """
        Load segmented images from training and testing directories
        
        Parameters:
        -----------
        training_dir : str
            Directory containing training images
        testing_dir : str
            Directory containing testing images
        max_samples_per_class : int or None
            Maximum number of samples to load per class (to limit dataset size)
            
        Returns:
        --------
        X_train, y_train, X_test, y_test : Training and testing data
        """
        X_train = []
        y_train = []
        X_test = []
        y_test = []
        
        # Load training data
        print("Loading training data...")
        for class_idx, class_name in enumerate(self.class_names):
            class_dir = os.path.join(training_dir, class_name)
            if not os.path.exists(class_dir):
                print(f"Warning: {class_dir} not found!")
                continue
            
            files = os.listdir(class_dir)
            # Filter segmented images
            segmented_files = [f for f in files if f.endswith("_segmented_result.png")]
            
            # Limit the number of samples if specified
            if max_samples_per_class:
                if len(segmented_files) > max_samples_per_class:
                    segmented_files = random.sample(segmented_files, max_samples_per_class)
                    print(f"Limited {class_name} training samples to {max_samples_per_class}")
            
            for file in tqdm(segmented_files):
                img_path = os.path.join(class_dir, file)
                img = cv2.imread(img_path)
                if img is None:
                    continue
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = cv2.resize(img, (self.img_width, self.img_height))
                img = img / 255.0  # Normalize to [0, 1]
                X_train.append(img)
                y_train.append(class_idx)
        
        # Load testing data
        print("Loading testing data...")
        for class_idx, class_name in enumerate(self.class_names):
            class_dir = os.path.join(testing_dir, class_name)
            if not os.path.exists(class_dir):
                print(f"Warning: {class_dir} not found!")
                continue
            
            files = os.listdir(class_dir)
            # Filter segmented images 
            segmented_files = [f for f in files if f.endswith("_segmented_result.png")]
            
            # Limit the number of samples if specified
            if max_samples_per_class:
                test_max = max(10, max_samples_per_class // 4)  # Ensure at least 10 test samples
                if len(segmented_files) > test_max:
                    segmented_files = random.sample(segmented_files, test_max)
                    print(f"Limited {class_name} testing samples to {test_max}")
            
            for file in tqdm(segmented_files):
                img_path = os.path.join(class_dir, file)
                img = cv2.imread(img_path)
                if img is None:
                    continue
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = cv2.resize(img, (self.img_width, self.img_height))
                img = img / 255.0  # Normalize to [0, 1]
                X_test.append(img)
                y_test.append(class_idx)
                    
        # Convert to numpy arrays
        X_train = np.array(X_train)
        y_train = np.array(y_train)
        X_test = np.array(X_test)
        y_test = np.array(y_test)
        
        # Convert labels to one-hot encoding
        y_train = to_categorical(y_train, num_classes=self.n_classes)
        y_test = to_categorical(y_test, num_classes=self.n_classes)
        
        print(f"Training data shape: {X_train.shape}")
        print(f"Training labels shape: {y_train.shape}")
        print(f"Testing data shape: {X_test.shape}")
        print(f"Testing labels shape: {y_test.shape}")
        
        # Check if we have enough data
        if len(X_train) < 10 or len(X_test) < 10:
            print("Warning: Very small dataset detected! Results may be unreliable.")
            
        return X_train, y_train, X_test, y_test
    
    def build_minimal_cnn_model(self):
        """
        Build a very lightweight CNN model for brain tumor classification
        """
        from tensorflow.keras.regularizers import l1, l2, l1_l2
        reg = l2(0.001)  # Adjust this value as needed (0.001 is a starting point)
    
        model = Sequential([
            # First convolutional block - add regularization to Conv2D layers
            Conv2D(16, (3, 3), activation='relu', padding='same', 
                kernel_regularizer=reg,
                input_shape=(self.img_height, self.img_width, 3)),
            MaxPooling2D(pool_size=(2, 2)),
            
            # Second convolutional block
            Conv2D(32, (3, 3), activation='relu', padding='same',
                kernel_regularizer=reg),
            MaxPooling2D(pool_size=(2, 2)),
            
            # Third convolutional block
            Conv2D(64, (3, 3), activation='relu', padding='same',
                kernel_regularizer=reg),
            MaxPooling2D(pool_size=(2, 2)),
            Dropout(0.3),  # Keep dropout as it's another form of regularization
            
            # Classification layers - add regularization to Dense layers
            Flatten(),
            Dense(128, activation='relu', kernel_regularizer=reg),
            Dropout(0.5),
            Dense(self.n_classes, activation='softmax')
        ])
        
        # Compile model
        model.compile(
            optimizer=Adam(learning_rate=0.0005),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        # Print model summary
        model.summary()
        
        self.model = model
        return model
    
    def create_simple_dataset(self, X, y, augment=False):
        """
        Create a batch generator for training (avoids using tf.data which can consume memory)
        
        Parameters:
        -----------
        X : np.array
            Image data
        y : np.array
            Labels
        augment : bool
            Whether to apply simple data augmentation
            
        Returns:
        --------
        generator function
        """
        n_samples = len(X)
        indices = np.arange(n_samples)
        
        # Define a generator function
        def batch_generator():
            while True:
                np.random.shuffle(indices)
                
                for start_idx in range(0, n_samples, self.batch_size):
                    batch_indices = indices[start_idx:start_idx + self.batch_size]
                    batch_X = X[batch_indices].copy()
                    batch_y = y[batch_indices].copy()
                    
                    # Simple augmentation if requested
                    if augment:
                        for i in range(len(batch_X)):
                            # Random horizontal flip with 50% probability
                            if np.random.rand() > 0.5:
                                batch_X[i] = np.fliplr(batch_X[i])
                    
                    yield batch_X, batch_y
                    
        # Calculate steps per epoch
        steps = int(np.ceil(n_samples / self.batch_size))
        
        return batch_generator(), steps
    
    def train(self, X_train, y_train, X_val=None, y_val=None, epochs=15, augment=True):
        """
        Train the CNN model with reduced complexity for CPU training
        
        Parameters:
        -----------
        X_train : np.array
            Training images
        y_train : np.array
            Training labels
        X_val : np.array
            Validation images
        y_val : np.array
            Validation labels
        epochs : int
            Number of training epochs (reduced from 50 to 15)
        augment : bool
            Whether to apply data augmentation
            
        Returns:
        --------
        history : Training history
        """
        if self.model is None:
            self.build_minimal_cnn_model()
            
        # If validation data not provided, split from training data
        if X_val is None or y_val is None:
            X_train, X_val, y_train, y_val = train_test_split(
                X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
            )
        
        # Create generators
        train_gen, train_steps = self.create_simple_dataset(X_train, y_train, augment=augment)
        val_gen, val_steps = self.create_simple_dataset(X_val, y_val, augment=False)
        
        # Define callbacks with more aggressive early stopping
        callbacks = [
            ModelCheckpoint(
                'brain_tumor_classifier_minimal.h5',
                save_best_only=True,
                monitor='val_accuracy',
                mode='max'
            ),
            EarlyStopping(
                patience=3,  # Very short patience to stop early
                monitor='val_loss',
                restore_best_weights=True
            ),
            ReduceLROnPlateau(
                factor=0.2,
                patience=2,  # Reduce learning rate quickly
                min_lr=1e-6,
                monitor='val_loss'
            )
        ]
        
        # Train model
        print("Training minimal CNN model...")
        history = self.model.fit(
            train_gen,
            steps_per_epoch=train_steps,
            validation_data=val_gen,
            validation_steps=val_steps,
            epochs=epochs,
            callbacks=callbacks
            # workers and use_multiprocessing parameters removed
        )
        
        self.history = history
        return history
        
    def evaluate(self, X_test, y_test):
        """
        Evaluate the trained model
        
        Parameters:
        -----------
        X_test : np.array
            Test images
        y_test : np.array
            Test labels
            
        Returns:
        --------
        results : dict
            Evaluation results
        """
        if self.model is None:
            print("Error: Model not trained yet!")
            return None
        
        # For CPU-friendly evaluation, use smaller batches
        batch_size = min(8, len(X_test))
        
        # Evaluate model
        print("Evaluating model...")
        loss, accuracy = self.model.evaluate(X_test, y_test, batch_size=batch_size, verbose=1)
        print(f"Test Loss: {loss}")
        print(f"Test Accuracy: {accuracy}")
        
        # Get predictions (in smaller batches to avoid memory issues)
        y_pred_all = []
        batch_size = min(16, len(X_test))
        for i in range(0, len(X_test), batch_size):
            batch_end = min(i + batch_size, len(X_test))
            batch_pred = self.model.predict(X_test[i:batch_end], verbose=0)
            y_pred_all.append(batch_pred)
        
        y_pred = np.vstack(y_pred_all)
        y_pred_classes = np.argmax(y_pred, axis=1)
        y_true_classes = np.argmax(y_test, axis=1)
        
        # Classification report
        print("\nClassification Report:")
        class_report = classification_report(
            y_true_classes, 
            y_pred_classes, 
            target_names=self.class_names,
            output_dict=True
        )
        print(classification_report(y_true_classes, y_pred_classes, target_names=self.class_names))
        
        # Confusion matrix
        print("\nConfusion Matrix:")
        conf_mat = confusion_matrix(y_true_classes, y_pred_classes)
        print(conf_mat)
        
        try:
            # Plot confusion matrix
            plt.figure(figsize=(10, 8))
            sns.heatmap(
                conf_mat, 
                annot=True, 
                fmt='d', 
                cmap='Blues', 
                xticklabels=self.class_names,
                yticklabels=self.class_names
            )
            plt.xlabel('Predicted')
            plt.ylabel('True')
            plt.title('Confusion Matrix')
            plt.tight_layout()
            plt.savefig('confusion_matrix.png')
            plt.show()
            
            # Plot training history if available
            if hasattr(self, 'history'):
                plt.figure(figsize=(12, 5))
                
                plt.subplot(1, 2, 1)
                plt.plot(self.history.history['accuracy'])
                plt.plot(self.history.history['val_accuracy'])
                plt.title('Model Accuracy')
                plt.ylabel('Accuracy')
                plt.xlabel('Epoch')
                plt.legend(['Train', 'Validation'], loc='upper left')
                
                plt.subplot(1, 2, 2)
                plt.plot(self.history.history['loss'])
                plt.plot(self.history.history['val_loss'])
                plt.title('Model Loss')
                plt.ylabel('Loss')
                plt.xlabel('Epoch')
                plt.legend(['Train', 'Validation'], loc='upper left')
                
                plt.tight_layout()
                plt.savefig('training_history.png')
                plt.show()
                
            # Sample predictions (only a few to save memory)
            self.visualize_predictions(X_test, y_true_classes, y_pred_classes, num_samples=3)
        except Exception as e:
            print(f"Warning: Could not create visualizations: {e}")
        
        return {
            'accuracy': accuracy,
            'loss': loss,
            'classification_report': class_report,
            'confusion_matrix': conf_mat
        }
    
    def visualize_predictions(self, X_test, y_true, y_pred, num_samples=3):
        """
        Visualize some sample predictions
        
        Parameters:
        -----------
        X_test : np.array
            Test images
        y_true : np.array
            True class indices
        y_pred : np.array
            Predicted class indices
        num_samples : int
            Number of samples to visualize (reduced to 3)
        """
        # Select random samples
        indices = np.random.choice(range(len(X_test)), size=min(num_samples, len(X_test)), replace=False)
        
        plt.figure(figsize=(12, 3 * num_samples))
        for i, idx in enumerate(indices):
            plt.subplot(num_samples, 1, i+1)
            plt.imshow(X_test[idx])
            if y_true[idx] == y_pred[idx]:
                color = 'green'
            else:
                color = 'red'
            plt.title(f"True: {self.class_names[y_true[idx]]}, Predicted: {self.class_names[y_pred[idx]]}", color=color)
            plt.axis('off')
        
        plt.tight_layout()
        plt.savefig('sample_predictions.png')
        plt.show()
    
    def predict(self, img_path):
        """
        Make prediction on a single image
        
        Parameters:
        -----------
        img_path : str
            Path to the image
            
        Returns:
        --------
        prediction : dict
            Prediction results including class and confidence
        """
        if self.model is None:
            print("Error: Model not trained yet!")
            return None
            
        # Read and preprocess image
        img = cv2.imread(img_path)
        if img is None:
            print(f"Error: Could not read image at {img_path}")
            return None
            
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (self.img_width, self.img_height))
        img = img / 255.0  # Normalize to [0, 1]
        
        # Make prediction
        pred = self.model.predict(np.expand_dims(img, axis=0))[0]
        pred_class = np.argmax(pred)
        
        try:
            # Show image and prediction
            plt.figure(figsize=(6, 6))
            plt.imshow(img)
            plt.title(f"Predicted: {self.class_names[pred_class]} ({pred[pred_class]:.2f})")
            plt.axis('off')
            plt.show()
        except Exception as e:
            print(f"Warning: Could not display image: {e}")
        
        return {
            'class': self.class_names[pred_class],
            'confidence': float(pred[pred_class]),
            'all_probabilities': {class_name: float(p) for class_name, p in zip(self.class_names, pred)}
        }
    
    def save_model(self, path='brain_tumor_classifier_minimal.h5'):
        """
        Save the trained model
        
        Parameters:
        -----------
        path : str
            Path to save the model
        """
        if self.model is None:
            print("Error: Model not trained yet!")
            return
            
        self.model.save(path)
        print(f"Model saved to {path}")
    
    def load_model(self, path='brain_tumor_classifier_minimal_v1_1.h5'):
        """
        Load a trained model
        
        Parameters:
        -----------
        path : str
            Path to the saved model
        """
        self.model = tf.keras.models.load_model(path)
        print(f"Model loaded from {path}")


def main():
    # Define paths to segmented data
    base_dir = "segmented_mri_dataset"  # Change this to your segmented data directory
    train_dir = os.path.join(base_dir, "Training")
    test_dir = os.path.join(base_dir, "Testing")
    
    # Initialize classifier with smaller image size
    classifier = BrainTumorClassifier(img_height=128, img_width=128, batch_size=16)
    
    # Load data with limit to reduce dataset size for CPU training
    # Adjust max_samples_per_class based on your computer's memory capacity
    max_samples = 300  # Limit samples per class to make training feasible on CPU
    X_train, y_train, X_test, y_test = classifier.load_data(
        train_dir, test_dir, max_samples_per_class=max_samples
    )
    
    # Build and train minimal CNN model
    classifier.build_minimal_cnn_model()
    history = classifier.train(X_train, y_train, epochs=50, augment=True)
    
    # Evaluate model
    results = classifier.evaluate(X_test, y_test)
    
    # Save model
    classifier.save_model()
    
    print("Done!")


if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.regularizers import l2
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.utils.class_weight import compute_class_weight
import cv2
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import random

# Set memory growth to avoid OOM errors
physical_devices = tf.config.list_physical_devices('GPU')
if physical_devices:
    for device in physical_devices:
        tf.config.experimental.set_memory_growth(device, True)
    print(f"GPU is available: {len(physical_devices)} devices found")
else:
    print("No GPU found, using CPU mode")
    # Set TensorFlow to use CPU only
    os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

# Set random seeds for reproducibility
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

class BrainTumorClassifier:
    def __init__(self, img_height=128, img_width=128, batch_size=8):
        """
        Initialize the Brain Tumor Classifier
        
        Parameters:
        -----------
        img_height : int
            Height to resize input images (reduced to 128x128)
        img_width : int
            Width to resize input images (reduced to 128x128)
        batch_size : int
            Batch size for training (reduced to 8)
        """
        self.img_height = img_height
        self.img_width = img_width
        self.batch_size = batch_size
        self.n_classes = 4  # glioma, meningioma, no_tumor, pituitary
        self.class_names = ['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']
        
        # Initialize model
        self.model = None
        
    def load_data(self, training_dir, testing_dir, max_samples_per_class=None):
        """
        Load segmented images from training and testing directories
        
        Parameters:
        -----------
        training_dir : str
            Directory containing training images
        testing_dir : str
            Directory containing testing images
        max_samples_per_class : int or None
            Maximum number of samples to load per class (to limit dataset size)
            
        Returns:
        --------
        X_train, y_train, X_test, y_test : Training and testing data
        """
        X_train = []
        y_train = []
        X_test = []
        y_test = []
        
        # Load training data
        print("Loading training data...")
        for class_idx, class_name in enumerate(self.class_names):
            class_dir = os.path.join(training_dir, class_name)
            if not os.path.exists(class_dir):
                print(f"Warning: {class_dir} not found!")
                continue
            
            files = os.listdir(class_dir)
            # Filter segmented images
            segmented_files = [f for f in files if f.endswith("_segmented_result.png")]
            
            # Limit the number of samples if specified
            if max_samples_per_class:
                if len(segmented_files) > max_samples_per_class:
                    segmented_files = random.sample(segmented_files, max_samples_per_class)
                    print(f"Limited {class_name} training samples to {max_samples_per_class}")
            
            for file in tqdm(segmented_files):
                img_path = os.path.join(class_dir, file)
                img = cv2.imread(img_path)
                if img is None:
                    continue
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = cv2.resize(img, (self.img_width, self.img_height))
                img = img / 255.0  # Normalize to [0, 1]
                X_train.append(img)
                y_train.append(class_idx)
        
        # Load testing data
        print("Loading testing data...")
        for class_idx, class_name in enumerate(self.class_names):
            class_dir = os.path.join(testing_dir, class_name)
            if not os.path.exists(class_dir):
                print(f"Warning: {class_dir} not found!")
                continue
            
            files = os.listdir(class_dir)
            # Filter segmented images 
            segmented_files = [f for f in files if f.endswith("_segmented_result.png")]
            
            # Limit the number of samples if specified
            if max_samples_per_class:
                test_max = max(10, max_samples_per_class // 4)  # Ensure at least 10 test samples
                if len(segmented_files) > test_max:
                    segmented_files = random.sample(segmented_files, test_max)
                    print(f"Limited {class_name} testing samples to {test_max}")
            
            for file in tqdm(segmented_files):
                img_path = os.path.join(class_dir, file)
                img = cv2.imread(img_path)
                if img is None:
                    continue
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = cv2.resize(img, (self.img_width, self.img_height))
                img = img / 255.0  # Normalize to [0, 1]
                X_test.append(img)
                y_test.append(class_idx)
                    
        # Convert to numpy arrays
        X_train = np.array(X_train)
        y_train = np.array(y_train)
        X_test = np.array(X_test)
        y_test = np.array(y_test)
        
        # Convert labels to one-hot encoding
        y_train = to_categorical(y_train, num_classes=self.n_classes)
        y_test = to_categorical(y_test, num_classes=self.n_classes)
        
        print(f"Training data shape: {X_train.shape}")
        print(f"Training labels shape: {y_train.shape}")
        print(f"Testing data shape: {X_test.shape}")
        print(f"Testing labels shape: {y_test.shape}")
        
        # Check if we have enough data
        if len(X_train) < 10 or len(X_test) < 10:
            print("Warning: Very small dataset detected! Results may be unreliable.")
        
        # Check class distribution
        self.check_class_distribution(y_train, y_test)
            
        return X_train, y_train, X_test, y_test
    
    def check_class_distribution(self, y_train, y_test):
        """
        Check distribution of classes in the dataset
        
        Parameters:
        -----------
        y_train : np.array
            Training labels (one-hot encoded)
        y_test : np.array
            Test labels (one-hot encoded)
        """
        # Convert one-hot encoded back to class indices
        train_classes = np.argmax(y_train, axis=1)
        test_classes = np.argmax(y_test, axis=1)
        
        print("\nClass Distribution:")
        # Count samples in each class
        for i, class_name in enumerate(self.class_names):
            train_count = np.sum(train_classes == i)
            test_count = np.sum(test_classes == i)
            print(f"{class_name}: {train_count} training, {test_count} testing samples")
    
    def visualize_training_samples(self, X, y, samples=5):
        """
        Visualize random training samples to check segmentation quality
        
        Parameters:
        -----------
        X : np.array
            Image data
        y : np.array
            Labels (one-hot encoded)
        samples : int
            Number of samples to visualize
        """
        y_indices = np.argmax(y, axis=1)
        plt.figure(figsize=(15, 3*samples))
        
        for i in range(samples):
            idx = np.random.randint(0, len(X))
            plt.subplot(samples, 1, i+1)
            plt.imshow(X[idx])
            plt.title(f"Class: {self.class_names[y_indices[idx]]}")
            plt.axis('off')
        
        plt.tight_layout()
        plt.savefig('training_samples.png')
        plt.show()
    
    def build_improved_cnn_model(self):
        """
        Build an improved CNN model for brain tumor classification
        """
        reg = l2(0.0005)  # Slightly reduced regularization
        
        model = Sequential([
            # First block
            Conv2D(32, (3, 3), activation='relu', padding='same', 
                kernel_regularizer=reg, input_shape=(self.img_height, self.img_width, 3)),
            Conv2D(32, (3, 3), activation='relu', padding='same', kernel_regularizer=reg),
            MaxPooling2D(pool_size=(2, 2)),
            Dropout(0.2),
            
            # Second block
            Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=reg),
            Conv2D(64, (3, 3), activation='relu', padding='same', kernel_regularizer=reg),
            MaxPooling2D(pool_size=(2, 2)),
            Dropout(0.3),
            
            # Third block
            Conv2D(128, (3, 3), activation='relu', padding='same', kernel_regularizer=reg),
            MaxPooling2D(pool_size=(2, 2)),
            Dropout(0.4),
            
            # Classification layers
            Flatten(),
            Dense(256, activation='relu', kernel_regularizer=reg),
            Dropout(0.5),
            Dense(self.n_classes, activation='softmax')
        ])
        
        # Compile with a slightly lower learning rate
        model.compile(
            optimizer=Adam(learning_rate=0.0001),  # Reduced learning rate
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        model.summary()
        self.model = model
        return model
    
    def build_transfer_learning_model(self):
        """
        Build a model based on pre-trained MobileNetV2
        
        Returns:
        --------
        model : Keras model
        """
        # Base model with pre-trained weights (excluding top layers)
        base_model = MobileNetV2(
            input_shape=(self.img_height, self.img_width, 3),
            include_top=False,
            weights='imagenet'
        )
        
        # Freeze the base model
        base_model.trainable = False
        
        # Add custom classification head
        model = Sequential([
            base_model,
            GlobalAveragePooling2D(),
            Dense(256, activation='relu'),
            Dropout(0.5),
            Dense(self.n_classes, activation='softmax')
        ])
        
        # Compile model
        model.compile(
            optimizer=Adam(learning_rate=0.0001),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        model.summary()
        self.model = model
        return model
    
    def create_advanced_dataset(self, X, y, augment=False):
        """
        Create a batch generator with more augmentation options
        
        Parameters:
        -----------
        X : np.array
            Image data
        y : np.array
            Labels
        augment : bool
            Whether to apply data augmentation
            
        Returns:
        --------
        generator function, steps per epoch
        """
        n_samples = len(X)
        indices = np.arange(n_samples)
        
        def batch_generator():
            while True:
                np.random.shuffle(indices)
                
                for start_idx in range(0, n_samples, self.batch_size):
                    batch_indices = indices[start_idx:start_idx + self.batch_size]
                    batch_X = X[batch_indices].copy()
                    batch_y = y[batch_indices].copy()
                    
                    # Enhanced augmentation
                    if augment:
                        for i in range(len(batch_X)):
                            img = batch_X[i]
                            
                            # Random horizontal flip (50% probability)
                            if np.random.rand() > 0.5:
                                img = np.fliplr(img)
                                
                            # Random vertical flip (20% probability)
                            if np.random.rand() > 0.8:
                                img = np.flipud(img)
                                
                            # Random rotation (0-15 degrees)
                            if np.random.rand() > 0.5:
                                angle = np.random.uniform(-15, 15)
                                rows, cols = img.shape[:2]
                                M = cv2.getRotationMatrix2D((cols/2, rows/2), angle, 1)
                                img = cv2.warpAffine(img, M, (cols, rows))
                            
                            # Random brightness/contrast adjustment
                            if np.random.rand() > 0.5:
                                # Adjust brightness
                                brightness = np.random.uniform(0.8, 1.2)
                                img = np.clip(img * brightness, 0, 1)
                            
                            batch_X[i] = img
                    
                    yield batch_X, batch_y
                        
        steps = int(np.ceil(n_samples / self.batch_size))
        return batch_generator(), steps
    
    def train(self, X_train, y_train, X_val=None, y_val=None, epochs=50, augment=True, use_class_weights=True):
        """
        Train the model with improved techniques
        
        Parameters:
        -----------
        X_train : np.array
            Training images
        y_train : np.array
            Training labels
        X_val : np.array
            Validation images
        y_val : np.array
            Validation labels
        epochs : int
            Number of training epochs
        augment : bool
            Whether to apply data augmentation
        use_class_weights : bool
            Whether to use class weights for imbalanced data
            
        Returns:
        --------
        history : Training history
        """
        if self.model is None:
            # By default, use transfer learning model which often performs better
            self.build_transfer_learning_model()
            
        # If validation data not provided, split from training data
        if X_val is None or y_val is None:
            X_train, X_val, y_train, y_val = train_test_split(
                X_train, y_train, test_size=0.2, random_state=42, stratify=np.argmax(y_train, axis=1)
            )
        
        # Create generators with advanced augmentation
        train_gen, train_steps = self.create_advanced_dataset(X_train, y_train, augment=augment)
        val_gen, val_steps = self.create_advanced_dataset(X_val, y_val, augment=False)
        
        # Compute class weights if needed
        class_weight_dict = None
        if use_class_weights:
            # Get class indices from one-hot encoded labels
            y_indices = np.argmax(y_train, axis=1)
            class_weights = compute_class_weight(
                'balanced', classes=np.unique(y_indices), y=y_indices
            )
            class_weight_dict = {i: w for i, w in enumerate(class_weights)}
            print("Using class weights:", class_weight_dict)
        
        # Define improved callbacks
        callbacks = [
            ModelCheckpoint(
                'brain_tumor_classifier_best.keras',
                save_best_only=True,
                monitor='val_accuracy',
                mode='max'
            ),
            EarlyStopping(
                patience=10,  # More patience
                monitor='val_loss',
                restore_best_weights=True
            ),
            ReduceLROnPlateau(
                factor=0.5,
                patience=5,
                min_lr=1e-7,
                monitor='val_loss'
            )
        ]
        
        # Train model
        print("Training model...")
        history = self.model.fit(
            train_gen,
            steps_per_epoch=train_steps,
            validation_data=val_gen,
            validation_steps=val_steps,
            epochs=epochs,
            callbacks=callbacks,
            # class_weight=class_weight_dict  # Apply class weights if available
        )
        
        self.history = history
        return history

    def cross_validate(self, X, y, n_folds=5, epochs=50):
        """
        Perform cross-validation training
        
        Parameters:
        -----------
        X : np.array
            Image data
        y : np.array
            Labels (one-hot encoded)
        n_folds : int
            Number of cross-validation folds
        epochs : int
            Number of epochs per fold
            
        Returns:
        --------
        fold_results : list
            Accuracy for each fold
        """
        # Convert one-hot back to class indices for stratification
        y_indices = np.argmax(y, axis=1)
        
        # Setup k-fold cross validation
        kfold = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=42)
        
        # Store results
        fold_results = []
        
        for fold, (train_idx, val_idx) in enumerate(kfold.split(X, y_indices)):
            print(f"\nTraining fold {fold+1}/{n_folds}")
            
            # Reset model for each fold
            self.model = None
            # Use transfer learning for better results
            self.build_transfer_learning_model()
            
            # Get fold data
            X_train_fold, X_val_fold = X[train_idx], X[val_idx]
            y_train_fold, y_val_fold = y[train_idx], y[val_idx]
            
            # Train on this fold
            history = self.train(
                X_train_fold, y_train_fold, 
                X_val=X_val_fold, y_val=y_val_fold,
                epochs=epochs, augment=True
            )
            
            # Evaluate on validation set
            results = self.evaluate(X_val_fold, y_val_fold)
            fold_results.append(results['accuracy'])
            
            print(f"Fold {fold+1} accuracy: {results['accuracy']:.4f}")
        
        print(f"\nCross-validation results:")
        print(f"Mean accuracy: {np.mean(fold_results):.4f}")
        print(f"Std deviation: {np.std(fold_results):.4f}")
        
        # Finally train on all data
        self.model = None
        self.build_transfer_learning_model()
        self.train(X, y, epochs=epochs, augment=True)
        
        return fold_results
        
    def evaluate(self, X_test, y_test):
        """
        Evaluate the trained model
        
        Parameters:
        -----------
        X_test : np.array
            Test images
        y_test : np.array
            Test labels
            
        Returns:
        --------
        results : dict
            Evaluation results
        """
        if self.model is None:
            print("Error: Model not trained yet!")
            return None
        
        # For CPU-friendly evaluation, use smaller batches
        batch_size = min(8, len(X_test))
        
        # Evaluate model
        print("Evaluating model...")
        loss, accuracy = self.model.evaluate(X_test, y_test, batch_size=batch_size, verbose=1)
        print(f"Test Loss: {loss}")
        print(f"Test Accuracy: {accuracy}")
        
        # Get predictions (in smaller batches to avoid memory issues)
        y_pred_all = []
        batch_size = min(16, len(X_test))
        for i in range(0, len(X_test), batch_size):
            batch_end = min(i + batch_size, len(X_test))
            batch_pred = self.model.predict(X_test[i:batch_end], verbose=0)
            y_pred_all.append(batch_pred)
        
        y_pred = np.vstack(y_pred_all)
        y_pred_classes = np.argmax(y_pred, axis=1)
        y_true_classes = np.argmax(y_test, axis=1)
        
        # Classification report
        print("\nClassification Report:")
        class_report = classification_report(
            y_true_classes, 
            y_pred_classes, 
            target_names=self.class_names,
            output_dict=True
        )
        print(classification_report(y_true_classes, y_pred_classes, target_names=self.class_names))
        
        # Confusion matrix
        print("\nConfusion Matrix:")
        conf_mat = confusion_matrix(y_true_classes, y_pred_classes)
        print(conf_mat)
        
        try:
            # Plot confusion matrix
            plt.figure(figsize=(10, 8))
            sns.heatmap(
                conf_mat, 
                annot=True, 
                fmt='d', 
                cmap='Blues', 
                xticklabels=self.class_names,
                yticklabels=self.class_names
            )
            plt.xlabel('Predicted')
            plt.ylabel('True')
            plt.title('Confusion Matrix')
            plt.tight_layout()
            plt.savefig('confusion_matrix.png')
            plt.show()
            
            # Plot training history if available
            if hasattr(self, 'history'):
                plt.figure(figsize=(12, 5))
                
                plt.subplot(1, 2, 1)
                plt.plot(self.history.history['accuracy'])
                plt.plot(self.history.history['val_accuracy'])
                plt.title('Model Accuracy')
                plt.ylabel('Accuracy')
                plt.xlabel('Epoch')
                plt.legend(['Train', 'Validation'], loc='upper left')
                
                plt.subplot(1, 2, 2)
                plt.plot(self.history.history['loss'])
                plt.plot(self.history.history['val_loss'])
                plt.title('Model Loss')
                plt.ylabel('Loss')
                plt.xlabel('Epoch')
                plt.legend(['Train', 'Validation'], loc='upper left')
                
                plt.tight_layout()
                plt.savefig('training_history.png')
                plt.show()
                
            # Sample predictions
            self.visualize_predictions(X_test, y_true_classes, y_pred_classes, num_samples=5)
        except Exception as e:
            print(f"Warning: Could not create visualizations: {e}")
        
        return {
            'accuracy': accuracy,
            'loss': loss,
            'classification_report': class_report,
            'confusion_matrix': conf_mat
        }
    
    def visualize_predictions(self, X_test, y_true, y_pred, num_samples=5):
        """
        Visualize some sample predictions
        
        Parameters:
        -----------
        X_test : np.array
            Test images
        y_true : np.array
            True class indices
        y_pred : np.array
            Predicted class indices
        num_samples : int
            Number of samples to visualize
        """
        # Select random samples
        indices = np.random.choice(range(len(X_test)), size=min(num_samples, len(X_test)), replace=False)
        
        plt.figure(figsize=(12, 3 * num_samples))
        for i, idx in enumerate(indices):
            plt.subplot(num_samples, 1, i+1)
            plt.imshow(X_test[idx])
            if y_true[idx] == y_pred[idx]:
                color = 'green'
            else:
                color = 'red'
            plt.title(f"True: {self.class_names[y_true[idx]]}, Predicted: {self.class_names[y_pred[idx]]}", color=color)
            plt.axis('off')
        
        plt.tight_layout()
        plt.savefig('sample_predictions.png')
        plt.show()
    
    def predict(self, img_path):
        """
        Make prediction on a single image
        
        Parameters:
        -----------
        img_path : str
            Path to the image
            
        Returns:
        --------
        prediction : dict
            Prediction results including class and confidence
        """
        if self.model is None:
            print("Error: Model not trained yet!")
            return None
            
        # Read and preprocess image
        img = cv2.imread(img_path)
        if img is None:
            print(f"Error: Could not read image at {img_path}")
            return None
            
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (self.img_width, self.img_height))
        img = img / 255.0  # Normalize to [0, 1]
        
        # Make prediction
        pred = self.model.predict(np.expand_dims(img, axis=0))[0]
        pred_class = np.argmax(pred)
        
        try:
            # Show image and prediction
            plt.figure(figsize=(6, 6))
            plt.imshow(img)
            plt.title(f"Predicted: {self.class_names[pred_class]} ({pred[pred_class]:.2f})")
            plt.axis('off')
            plt.show()
        except Exception as e:
            print(f"Warning: Could not display image: {e}")
        
        return {
            'class': self.class_names[pred_class],
            'confidence': float(pred[pred_class]),
            'all_probabilities': {class_name: float(p) for class_name, p in zip(self.class_names, pred)}
        }
    
    def save_model(self, path='brain_tumor_classifier_best.keras'):
        """
        Save the trained model
        
        Parameters:
        -----------
        path : str
            Path to save the model
        """
        if self.model is None:
            print("Error: Model not trained yet!")
            return
            
        self.model.save(path)
        print(f"Model saved to {path}")
    
    def load_model(self, path='brain_tumor_classifier_best.keras'):
        """
        Load a trained model
        
        Parameters:
        -----------
        path : str
            Path to the saved model
        """
        self.model = tf.keras.models.load_model(path)
        print(f"Model loaded from {path}")


def main():
    # Define paths to segmented data
    base_dir = "/kaggle/input/mri-images/segmented_mri_dataset"  # Change this to your segmented data directory
    train_dir = os.path.join(base_dir, "Training")
    test_dir = os.path.join(base_dir, "Testing")
    
    # Initialize classifier with optimal parameters
    classifier = BrainTumorClassifier(img_height=224, img_width=224, batch_size=16)
    
    # Load data with appropriate sample size for training
    max_samples = 500  # Increase if you have more memory available
    X_train, y_train, X_test, y_test = classifier.load_data(
        train_dir, test_dir, max_samples_per_class=max_samples
    )
    
    # Visualize some training samples to check quality
    classifier.visualize_training_samples(X_train, y_train, samples=3)
    
    # Option 1: Use transfer learning (recommended for better accuracy)
    classifier.build_transfer_learning_model()
    history = classifier.train(X_train, y_train, epochs=100, augment=True, use_class_weights=True)
    
    # Option 2: Use cross-validation for even better results (but slower)
    # fold_results = classifier.cross_validate(np.vstack((X_train, X_test)), 
    #                                         np.vstack((y_train, y_test)), 
    #                                         n_folds=5, epochs=30)
    
    # Evaluate model
    results = classifier.evaluate(X_test, y_test)
    
    # Save the best model
    classifier.save_model('brain_tumor_classifier_improved.keras')
    
    print("Training and evaluation completed!")


if __name__ == "__main__":
    main()
```
